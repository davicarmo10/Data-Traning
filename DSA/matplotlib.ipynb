```python
#Gráfico de linha

#Dados
x = np.linspace(0, 5, 11)
y = x ** 2  


#Cria a figura
fig = plt.figure()

#define a escala dos eixos  
axes1 = fig.add_axes([0.5, 0.5, 0.5, 0.5])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

#Cria o plot
axes1.plot(x, y, 'r')
axes2.plot(y, x, 'g')

#Labels e títulos
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('title')

axes2.plot(y, x, 'g')
axes2.set_ylabel = 'y'
axes2.set_xlabel = 'x'
axes2.set_title = 'title2'  

plt.show()
```

```python
#import numpy as np
#import matplotlib.pyplot as plt

# Definição dos dados
x = np.linspace(0, 5, 10)  # Gera 10 pontos entre 0 e 5
y1 = x**2  # Quadrado de x
y2 = np.exp(x)  # Exponencial de x

# Criação dos subplots
fig, axes = plt.subplots(1, 4, figsize=(10, 4))

# Primeiro gráfico - Escala padrão
axes[0].plot(x, x * 2, label="y = 2x")
axes[0].plot(x, y2, label="y = e^x")
axes[0].set_title("Escala Padrão")
axes[0].legend()

# Segundo gráfico - Escala logarítmica no eixo Y
axes[1].plot(x, y1, label="y = x²")
axes[1].plot(x, y2, label="y = e^x")
axes[1].set_yscale("log")  # Define escala logarítmica para o eixo Y
axes[1].set_title("Escala Logarítmica (y)")
axes[1].grid(color = 'b', linestyle = '--', linewidth = 0.5)
axes[1].legend()

axes[2].scatter(x, y1, label="y = x²")

axes[3].bar(x, y2, label="y = x²")

# Exibe os gráficos
plt.tight_layout()
plt.show()
```
